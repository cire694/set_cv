## Testing the contour finding

```python
import extract_cards
```

```python
import cv2 as cv
import matplotlib.pyplot as plt
def show_contours(path):
    img_copy = cv.imread(path).copy()
    contours = extract_cards.get_card_contours(path)
    img_copy = cv.drawContours(img_copy, contours, -1, (0, 255, 0), 2)
    
    img_rbg = cv.cvtColor(img_copy, cv.COLOR_BGR2RGB)

    plt.figure(figsize=(10, 10))
    plt.imshow(img_rbg)
    plt.axis("off")
    plt.tight_layout()
    plt.show()
```

```python
path = "../images/set.png"
show_contours(path)
```
